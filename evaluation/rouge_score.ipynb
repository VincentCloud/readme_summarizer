# Objective
Calculate rouge score for each datapoint on desired columns


## Import Necessary Modules and Packages



```python
import pandas as pd
import numpy as np
from pyrouge import Rouge155
import os

REFERENCE_PATH = '/Users/vincenthuang/Development/Summer-2020/readme_summarizer/data/output/reference'
HYPOTHESIS_PATH = '/Users/vincenthuang/Development/Summer-2020/readme_summarizer/data/output/hypothesis'
REFERENCE_FILE = 'ref.1.txt'
HYPOTHESIS_FILE = 'hyp.A.1.txt'
```

## Calculate ROUGE Score for One Sample



### Write the results to .txt files

```python
# One file
def write_to_txt_files(path, file_name, content):
    with open(f'{path}/{file_name}', 'w+') as fn:
        fn.write(content)
    return content
```

```python
# Delete a file
def remove_file(path, file_name):
    os.remove(f'{path}/{file_name}')
```

### Calculate ROUGE Score for One Data Point

```python
def calculate_rouge_score(system_dir, model_dir):
    r = Rouge155()
    r.system_dir = system_dir
    r.model_dir = model_dir
    r.system_filename_pattern = 'ref.(\d+).txt'
    r.model_filename_pattern = f'hyp.A.#ID#.txt'

    output = r.convert_and_evaluate()
    output_dict = r.output_to_dict(output)
    return output, output_dict

# Parse ROUGE score and take out fn
def parse_rouge_score(output_dict):
    fn_score = (output_dict['rouge_1_f_score'] + output_dict['rouge_2_f_score'] + output_dict['rouge_l_f_score']) / 3
    return fn_score

```

## Read from Data

```python
def read_from_csv(path):
    return pd.read_csv(path)

def get_column(df,column):
    return df[column]
```

```python
# Calculate rouge score for two columns
def calculate_individual_rouge_score(source_path, csv_file, ref_col, hyp_col):
    df = read_from_csv(f'{source_path}/{csv_file}').drop('Unnamed: 0', 1)
    refs = df[ref_col]
    hyps = df[hyp_col]
    scores = []
    for i in range(len(hyps)):
        write_to_txt_files(REFERENCE_PATH, REFERENCE_FILE, refs[i])
        write_to_txt_files(HYPOTHESIS_PATH, HYPOTHESIS_FILE, hyps[i])

        _, output_dict = calculate_rouge_score(REFERENCE_PATH, HYPOTHESIS_PATH)
        score = parse_rouge_score(output_dict)

        scores.append(score)

        remove_file(REFERENCE_PATH, REFERENCE_FILE)
        remove_file(HYPOTHESIS_PATH, HYPOTHESIS_FILE)

    df[f'{hyp_col} score'] = scores
    df.to_csv(f'{source_path}/{hyp_col}_scores.csv')
    return df
```
